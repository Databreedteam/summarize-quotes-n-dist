```python
df_train = df_raw[df_raw["record_type"] == 1]
df_train["A-G"] = df_train["A"].astype(str) + df_train["B"].astype(str) + df_train["C"].astype(str) + df_train["D"].astype(str) + df_train["E"].astype(str) + df_train["F"].astype(str) + df_train["G"].astype(str)
print("Rows by Columns: ", df_train.shape)
print("Number of Unique Levels in A-G: ", np.shape(df_train["A-G"].unique())[0])
```

```python
df_quotes = df_raw[df_raw["record_type"] == 0]
df_quotes.head(20)
```

```python
from scipy.spatial import distance

def computeEUDistance(grp_df):
    cumm_dist = 0
    dist = 0
    for i in range(grp_df.shape[0]):
        if (i+1 <= (grp_df.shape[0] - 1)):
            # print(grp_df.iloc[i+1, 17], ":", grp_df.iloc[i+1, 18])
            a = (grp_df.iloc[i+1, 17].astype(int), grp_df.iloc[i+1, 18].astype(int), grp_df.iloc[i+1, 19].astype(int), grp_df.iloc[i+1, 20].astype(int), grp_df.iloc[i+1, 21].astype(int), grp_df.iloc[i+1, 22].astype(int), grp_df.iloc[i+1, 23].astype(int))
            b = (grp_df.iloc[i, 17].astype(int), grp_df.iloc[i, 18].astype(int), grp_df.iloc[i, 19].astype(int), grp_df.iloc[i, 20].astype(int), grp_df.iloc[i, 21].astype(int), grp_df.iloc[i, 22].astype(int), grp_df.iloc[i, 23].astype(int))
            dist = distance.euclidean(a, b)
        cumm_dist += dist
    return cumm_dist

def computeNorminalDistance(grp_df):
    cumm_dist = 0
    dist = 0
    for i in range(grp_df.shape[0]):
        if (i+1 <= (grp_df.shape[0] - 1)):
            a = grp_df.iloc[i, 17] == grp_df.iloc[i+1, 17]
            b = grp_df.iloc[i, 18] == grp_df.iloc[i+1, 18]
            c = grp_df.iloc[i, 19] == grp_df.iloc[i+1, 19]
            d = grp_df.iloc[i, 20] == grp_df.iloc[i+1, 20]
            e = grp_df.iloc[i, 21] == grp_df.iloc[i+1, 21]
            f = grp_df.iloc[i, 22] == grp_df.iloc[i+1, 22]
            g = grp_df.iloc[i, 23] == grp_df.iloc[i+1, 23]
            # print(str(int(a) + int(b) + int(c) + int(d) + int(e) + int(f) + int(g)))
            dist = (7 - (int(a) + int(b) + int(c) + int(d) + int(e) + int(f) + int(g))) / 7
        cumm_dist += dist
    return cumm_dist

# Select top 100 quotes for testing
df_top100 = df_quotes.iloc[0:100, :]

# Compute norminal distance using (p - m)/p
resN = df_quotes.groupby("customer_ID").apply(computeNorminalDistance)
resN.name = "N_DIST"
#print(resN)

# Compute euclidean distance assuming A to G values are numerical
resEU = df_quotes.groupby("customer_ID").apply(computeEUDistance)
resEU.name = "EU_DIST"
#print(resEU)

df_comb = pd.DataFrame(data = dict(N_DIST = resN, EU_DIST = resEU))
print(df_comb.head())
```
